# Exercise 3.2: Split-Apply-Combine of the frog data set

```python
import numpy as np
import pandas as pd
```

```python
ID_grouped = df.groupby('ID')

ID_grouped
```

```python
IF_SD = ID_grouped['impact force (mN)'].std(numeric_only=True)
```

```python
IF_SD = df.groupby('ID')['impact force (mN)'].std(numeric_only=True)
```

```python
IF_SD
```

```python
def coeff_of_var(data):
    return np.std(data) / np.mean(data)
```

```python
IF_AF_CV = ID_grouped[['adhesive force (mN)', 'impact force (mN)']].agg(coeff_of_var).reset_index()
```

```python
IF_AF_CV
```

```python
df_IF_AF = (
    df.groupby('ID')[['adhesive force (mN)', 'impact force (mN)']]
    .agg(['mean', 'median', 'std', coeff_of_var])
    .reset_index()
)
```

```python
df_IF_AF
```

```python
df_tidy = pd.melt(df_IF_AF, var_name=['force_type', 'stat_measure'], id_vars='ID')
df_tidy.head(20)
```

# Exercise 3.3: Adding data to a data frame

```python
df_frog = pd.DataFrame(
    data={
        "ID": ["I", "II", "III", "IV"],
        "age": ["adult", "adult", "juvenile", "juvenile"],
        "SVL (mm)": [63, 70, 28, 31],
        "weight (g)": [63.1, 72.7, 12.7, 12.7],
        "species": ["cross", "cross", "cranwelli", "cranwelli"],
    }
)
```

```python
df_frog
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')
```

```python
df_new = df.merge(df_frog, on = 'ID')
```

```python
df_new
```

```python

df['max_IF'] = df.groupby('ID')['impact force (mN)'].transform('max').eq(df["impact force (mN)"])

def highlight_max_IF(row):
    if row['max_IF']:
        return ['background-color: #99d8c9'] * len(row)
    else:
        return ['background-color: lightgray'] * len(row)

styled_df = df.style.apply(highlight_max_IF, axis=1)
```
