```python
# Largest number to consider
n_max = 1000

# Initialize list of primes
primes = []

# Loop through odd integers and add primes
for x in range(n_max+1): # +1 to include 1000 itself
    if sympy.isprime(x):
        primes.append(x)

# Take a look
print(primes)
```

```python
primes = [x for x in range(n_max) if sympy.isprime(x)]

print(primes)
```

```python
def coeff_of_var(data):
    return np.std(data) / np.mean(data)
```

```python
cols = df.columns[
      (df.columns != 'gender') 
    & (df.columns != 'insomnia')
    & (df.columns != 'participant number')
]

df.groupby('insomnia')['percent correct'].agg(coeff_of_var).reset_index()
```

```python
def insom_scale(x):
    if x <= 16:
        return 'insomniac'
    elif x<= 25:
        return 'restless'
    else:
        return 'restful'

df['insom_scale'] = df['sci'].apply(insom_scale)
```

```python
df.groupby(['gender', 'insom_scale'])['percent correct'].mean().reset_index()
```

```python
import pandas as pd

# Our main plotting package (must have explicit import of submodules)
import bokeh.io
import bokeh.models
import bokeh.plotting

# Enable viewing Bokeh plots in the notebook
bokeh.io.output_notebook()
```

```python
# Create the figure, stored in variable `p`
p = bokeh.plotting.figure(
    frame_width=300,
    frame_height=300,
    x_axis_label='confidence when correct',
    y_axis_label='condifence when incorrect',
    x_range=[-5, 105],
    y_range=[-5, 105],
)
```

```python
p.circle(
    source=df,
    x='confidence when correct',
    y='confidence when incorrect'
);
```

```python
bokeh.io.show(p)
```

```python
# For convenience
x = 'confidence when correct'
y = 'confidence when incorrect'

# Two sources
df_insomnia = df.loc[df['insomnia'], :]
df_normal = df.loc[~df['insomnia'], :]

# Make figure
p = bokeh.plotting.figure(
    frame_width=300,
    frame_height=300,
    x_axis_label=x,
    y_axis_label=y,
    x_range=[-5, 105],
    y_range=[-5, 105],
    
)

# Add glyphs
p.circle(
    source=df_normal,
    x=x,
    y=y,
    legend_label='normal sleepers',
)

p.circle(
    source=df_insomnia,
    x=x,
    y=y,
    color='orange',
    legend_label='insomniacs',
)

p.legend.location = 'top_left'

bokeh.io.show(p)
```

```python
import numpy as np
import pandas as pd

import iqplot

import bokeh.io
import bokeh.models
import bokeh.plotting

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv', comment='#')

# Have a look so we remember
df.head()
```

```python
df_mean = df.groupby('ID')['impact force (mN)'].mean().reset_index()

# Take a look
df_mean
```

```python
p = bokeh.plotting.figure(
    frame_height=200,
    frame_width=400,
    x_axis_label='impact force (mN)',
    y_range=df_mean['ID'].unique()[::-1],
    tools='pan,wheel_zoom,save,reset'
)
```

```python
p = iqplot.box(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.strip(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.stripbox(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.striphistogram(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.ecdf(
    data=df,
    q="impact force (mN)",
    cats="ID",
    style="staircase",
)

bokeh.io.show(p)
```
